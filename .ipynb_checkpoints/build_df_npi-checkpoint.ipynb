```python
# import df vertical
df_vertical = src.read_from_efs('df_final_vertical.csv')
src.print_info(df_vertical, 'df_vertical')
df_vertical.tail()
```

```python
df = df_vertical.groupby('npi').agg('sum')
df.drop(columns=['cost_per_bene','cost_per_claim','label'], inplace=True)
df['cost_per_bene'] = df['total_drug_cost']/df['bene_count']
df['cost_per_claim'] = df['total_drug_cost']/df['total_claim_count']
src.print_info(df, 'df')
df.tail()
```
